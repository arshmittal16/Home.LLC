```python
import pandas as pd
import matplotlib.pyplot as plt
# %matplotlib inline
import seaborn as sns
import numpy as np
from scipy import stats
```

```python
prices = pd.read_csv('HPI.csv')
unemp = pd.read_csv('file.csv')
gdp = pd.read_csv('GDP.csv')
cpi = pd.read_csv('cpi.csv')
mort = pd.read_csv('MORTGAGE30US.csv')
csr = pd.read_csv('MSACSR.csv')
```

```python
print(prices.head())
print(gdp.head())
print(csr.head())
print(mortgage.head())
print(unemp.head())
print(cpi.head())
```

```python
final = {'DATE': prices.iloc[:253,0],
        'Unemployment': unemp.iloc[:253, 1],
        'GDP Growth Rate': gdp.iloc[:253, 1],
        'Consumer Price Index': cpi.iloc[:253, 1],
        'Mortgage Rate': mortgage.iloc[:253, 1],
        'Supply of New Homes': csr.iloc[:253, 1],
        'Home Price Index': prices.iloc[:253, 1]}

df = pd.DataFrame(final)
df.head()
```

```python
df.head()
```

The DataFrame 'ds' is the entire dataset clubbed with all the factors that affected the home prices in the last 20 years (2000 - 2020) turned into specific columns.

```python
## Relationship between Unemployment & Home Prices
figure, axis = plt.subplots(1, 2)
x = df.iloc[:, 0]
y = df.iloc[:, -1]
z = df.iloc[:, 1]

axis[0].plot(x,y)
axis[0].set_title('Home Prices over time')

axis[1].plot(x, z)
axis[1].set_title('Unemployment over time')

plt.show()
```

As we can see, Unemployment isn't exactly affecting the Home Prices but most of the time in the last 20 years, it has been indirectly proportional to the prices.

```python
## Relationship between GDP Growth Rate & Home Prices
figure, axis = plt.subplots(1, 2)
x = df.iloc[:252, 0]
y = df.iloc[:252, -1]
z = df.iloc[:252, 2]

axis[0].plot(x,y)
axis[0].set_title('Home Prices over time')

axis[1].plot(x, z)
axis[1].set_title('GDP Growth over time')

plt.show()
```

GDP is a clear indicator of Home Price Index. In the last 20 years(except when The Great Recession happened), Home Prices have been directly proportional to the GDP of United States. 

```python
## Relationship between Consumer Price Index & Home Prices
figure, axis = plt.subplots(1, 2)
x = df.iloc[:252, 0]
y = df.iloc[:252, -1]
z = df.iloc[:252, 3]

axis[0].plot(x,y)
axis[0].set_title('Home Prices over time')

axis[1].plot(x, z)
axis[1].set_title('Consumer Price Index over time')

plt.show()
```

Consumer Price Index has always changed in accordance to Home Price Index in the last 20 years. Their graph is almost identical and hence they both are directly proportional to each other.

```python
## Relationship between Mortgage Rate & Home Prices
figure, axis = plt.subplots(1, 2)
x = df.iloc[:252, 0]
y = df.iloc[:252, -1]
z = df.iloc[:252, 4]

axis[0].plot(x,y)
axis[0].set_title('Home Prices over time')

axis[1].plot(x, z)
axis[1].set_title('Mortgage Rate over time')

plt.show()
```

Mortgage Rates play a major role in determining Home Prices. When the Mortgage Rates were high, demand was low and hence the prices. But, when the rates went lower, demand went up and hence the prices. Therefore, as we can see, Home Prices and Mortgage Rates have always been inversely proportional to each other.

```python
## Relationship between Supply of New Homes & Home Prices
figure, axis = plt.subplots(1, 2)
x = df.iloc[:252, 0]
y = df.iloc[:252, -1]
z = df.iloc[:252, -2]

axis[0].plot(x,y)
axis[0].set_title('Home Prices over time')

axis[1].plot(x, z)
axis[1].set_title('Supply of new homes')

plt.show()
```

Supply of new houses is the only factor in this dataset which contributes to the Supply part of Home Prices. As we can see, in the earlier part of bi-decade supply increased every year and so did the prices, meaning number of buyers were much more than the number of new homes built. Whereas, in the later part of bi-decade, supply went down and the prices still went up, meaning demand was constant but the supply was down, hence increase in prices.

```python
print(df.corr().iloc[:, -1])
```

Compare_Models tells us the best model to create if we want to design a predictive model for the home prices based on the input features, namely Unemployment,GDP Growth Rate,Consumer Price Index,Mortgage Rate and Supply of New Homes.

Since, we now know Random Forest will give us the best results, we will use RandomForestRegressor to check which features/factors have had the highest and the lowest impact on Home Prices over the last 20 years.

```python
from sklearn.ensemble import RandomForestRegressor

X = df.iloc[:, 1:6]
y = df.iloc[:, 6]

model = RandomForestRegressor()

model.fit(X, y)
```

```python
importance = model.feature_importances_
```

```python
for i,v in enumerate(importance):
    print('Feature: %0d, Score: %.5f' % (i+1,v))
```

```python
plt.bar([x for x in range(1, len(importance)+1)], importance)
plt.show()
```

### Conclusion: 'GDP Growth' had the highest impact on Home Prices in the last 20 years, while 'Supply of New Homes' had the lowest impact on Home Prices.
